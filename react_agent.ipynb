```python
# ==============================
# Minimal ReAct Infra Agent
# ==============================

import json

# ------------------------------
# Mock Tools
# ------------------------------

def check_disk():
    return {
        "disk_usage": "95%",
        "status": "critical"
    }

def get_pod_status():
    return {
        "pod_name": "payment-service",
        "status": "CrashLoopBackOff"
    }

def check_cpu():
    return {
        "cpu_usage": "92%",
        "status": "high"
    }

# ------------------------------
# ReAct Agent
# ------------------------------

class ReActAgent:

    def run(self, issue):

        trace = []

        # Thought
        thought1 = f"The system reported: {issue}. I should inspect infrastructure metrics."
        trace.append(("Thought", thought1))

        # Action
        action1 = "check_cpu()"
        trace.append(("Action", action1))

        observation1 = check_cpu()
        trace.append(("Observation", observation1))

        # Reasoning
        thought2 = "CPU usage is high. This may affect application pods. I should check pod health."
        trace.append(("Thought", thought2))

        # Action
        action2 = "get_pod_status()"
        trace.append(("Action", action2))

        observation2 = get_pod_status()
        trace.append(("Observation", observation2))

        # Reasoning
        thought3 = "Pod is crashing repeatedly. Disk pressure may also be involved. I should inspect disk usage."
        trace.append(("Thought", thought3))

        # Action
        action3 = "check_disk()"
        trace.append(("Action", action3))

        observation3 = check_disk()
        trace.append(("Observation", observation3))

        # Final Answer
        answer = {
            "incident": "Application Pod Crash",
            "severity": "High",
            "root_cause": "High CPU and critical disk usage caused pod instability.",
            "affected_service": observation2["pod_name"],
            "recommended_action": [
                "Clean unused disk files",
                "Increase node storage",
                "Restart affected pod",
                "Monitor CPU metrics"
            ]
        }

        trace.append(("Answer", answer))

        return trace


# ------------------------------
# Run Agent
# ------------------------------

agent = ReActAgent()

result = agent.run("Users report service downtime")

# Print ReAct Trace
for step_type, content in result:
    print(f"\n{step_type}:")
    print(content)
```

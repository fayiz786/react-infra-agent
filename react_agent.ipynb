```python
# ============================================
# Advanced Minimal ReAct Infra Agent
# Multiple Issue Simulation + Colored Logs
# ============================================

import json
import random
from datetime import datetime

# ANSI Colors
class Colors:
    HEADER = '\033[95m'
    BLUE = '\033[94m'
    CYAN = '\033[96m'
    GREEN = '\033[92m'
    WARNING = '\033[93m'
    RED = '\033[91m'
    BOLD = '\033[1m'
    END = '\033[0m'


# ============================================
# Mock Tools
# ============================================

def check_cpu():
    cpu = random.randint(70, 99)

    return {
        "cpu_usage": f"{cpu}%",
        "status": "high" if cpu > 85 else "normal"
    }


def check_disk():
    disk = random.randint(75, 99)

    return {
        "disk_usage": f"{disk}%",
        "status": "critical" if disk > 90 else "warning"
    }


def get_pod_status():
    pod_states = [
        "Running",
        "CrashLoopBackOff",
        "ImagePullBackOff",
        "OOMKilled"
    ]

    return {
        "pod_name": "payment-service",
        "status": random.choice(pod_states)
    }


def check_memory():
    memory = random.randint(65, 98)

    return {
        "memory_usage": f"{memory}%",
        "status": "high" if memory > 85 else "normal"
    }


# ============================================
# ReAct Agent
# ============================================

class ReActAgent:

    def log(self, step, content, color):

        print(f"\n{color}{'=' * 60}")
        print(f"{step}")
        print(f"{'=' * 60}{Colors.END}")

        print(content)

    def run(self, issue):

        trace = []

        # ---------------------------
        # THOUGHT 1
        # ---------------------------

        thought1 = (
            f"The system reported: '{issue}'. "
            f"I should inspect CPU utilization first."
        )

        trace.append(("Thought", thought1))
        self.log("THOUGHT", thought1, Colors.CYAN)

        # ---------------------------
        # ACTION 1
        # ---------------------------

        action1 = "check_cpu()"

        trace.append(("Action", action1))
        self.log("ACTION", action1, Colors.BLUE)

        observation1 = check_cpu()

        trace.append(("Observation", observation1))
        self.log("OBSERVATION", observation1, Colors.WARNING)

        # ---------------------------
        # THOUGHT 2
        # ---------------------------

        thought2 = (
            "CPU usage seems abnormal. "
            "I should verify Kubernetes pod health."
        )

        trace.append(("Thought", thought2))
        self.log("THOUGHT", thought2, Colors.CYAN)

        # ---------------------------
        # ACTION 2
        # ---------------------------

        action2 = "get_pod_status()"

        trace.append(("Action", action2))
        self.log("ACTION", action2, Colors.BLUE)

        observation2 = get_pod_status()

        trace.append(("Observation", observation2))
        self.log("OBSERVATION", observation2, Colors.WARNING)

        # ---------------------------
        # THOUGHT 3
        # ---------------------------

        thought3 = (
            "Pod instability detected. "
            "Disk pressure could be contributing."
        )

        trace.append(("Thought", thought3))
        self.log("THOUGHT", thought3, Colors.CYAN)

        # ---------------------------
        # ACTION 3
        # ---------------------------

        action3 = "check_disk()"

        trace.append(("Action", action3))
        self.log("ACTION", action3, Colors.BLUE)

        observation3 = check_disk()

        trace.append(("Observation", observation3))
        self.log("OBSERVATION", observation3, Colors.WARNING)

        # ---------------------------
        # THOUGHT 4
        # ---------------------------

        thought4 = (
            "Memory exhaustion may also affect pod stability. "
            "Checking memory metrics."
        )

        trace.append(("Thought", thought4))
        self.log("THOUGHT", thought4, Colors.CYAN)

        # ---------------------------
        # ACTION 4
        # ---------------------------

        action4 = "check_memory()"

        trace.append(("Action", action4))
        self.log("ACTION", action4, Colors.BLUE)

        observation4 = check_memory()

        trace.append(("Observation", observation4))
        self.log("OBSERVATION", observation4, Colors.WARNING)

        # ============================================
        # INCIDENT ANALYSIS
        # ============================================

        issues = []

        if observation1["status"] == "high":
            issues.append("High CPU Usage")

        if observation3["status"] == "critical":
            issues.append("Critical Disk Usage")

        if observation4["status"] == "high":
            issues.append("High Memory Usage")

        if observation2["status"] != "Running":
            issues.append(f"Pod Failure ({observation2['status']})")

        # ============================================
        # FINAL INCIDENT REPORT
        # ============================================

        answer = {
            "timestamp": str(datetime.now()),
            "incident": "Infrastructure Health Degradation",
            "severity": "High",
            "detected_issues": issues,
            "affected_service": observation2["pod_name"],
            "root_cause": (
                "Infrastructure resource exhaustion caused "
                "application instability."
            ),
            "recommended_actions": [
                "Clean unused disk files",
                "Scale Kubernetes nodes",
                "Restart failing pods",
                "Optimize application resource usage",
                "Enable continuous monitoring alerts"
            ]
        }

        trace.append(("Answer", answer))

        self.log(
            "FINAL INCIDENT REPORT",
            json.dumps(answer, indent=4),
            Colors.GREEN
        )

        return trace


# ============================================
# RUN AGENT
# ============================================

if __name__ == "__main__":

    agent = ReActAgent()

    result = agent.run(
        "Users report payment API downtime and slow response."
    )
```
